```python
import pandas as pd
data = pd.read_csv("nuevo.csv")
data = data.dropna()
data.to_csv('data_new.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler

# Cargar el dataset de Energia
df_gas = pd.read_csv("data_new.csv")
cols = ["gas_prod_change_twh", "gas_electricity", "gas_share_energy", "gas_cons_change_twh", "gas_consumption", "gas_production", "Natural Gas /US dollars per million Btu", "Annual CO2 emissions"]
df_gas = df_gas[cols]

# Variables independientes (X)
X = df_gas[["gas_share_energy", "gas_cons_change_twh", "gas_consumption", "Natural Gas /US dollars per million Btu", "Annual CO2 emissions"]]
# Variable dependiente (y)
y = df_gas["gas_production"].values

# Dividir el dataset en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=0)

# Escalar los datos de prueba
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crear y entrenar el modelo de Regresi√≥n Lineal M√∫ltiple
regressor = LinearRegression()
regressor.fit(X_train_scaled, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = regressor.predict(X_test_scaled)

# Calcular el error cuadr√°tico medio (MSE) en el conjunto de prueba
mse = mean_squared_error(y_test, y_pred)
print("Error cuadr√°tico medio (MSE):", mse)

# Graficar los valores reales y las predicciones
plt.plot(y_test, label='Valores reales')
plt.plot(y_pred, label='Predicciones')
plt.xlabel('√çndice de muestra')
plt.ylabel('Producci√≥n de gas')
plt.title('Predicci√≥n de Producci√≥n de Gas Natural')
plt.legend()
plt.show()
```
